```python
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt

# Load datasets
train_df = pd.read_csv("/content/train_set.csv")
val_df = pd.read_csv("/content/val_set.csv")
test_df = pd.read_csv("/content/test_set.csv")

# Select numerical feature
X_train = np.array(train_df['num_values']).reshape(-1, 1)
X_val = np.array(val_df['num_values']).reshape(-1, 1)
X_test = np.array(test_df['num_values']).reshape(-1, 1)

# Normalize data
scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_val = scaler.transform(X_val)
X_test = scaler.transform(X_test)

# Reshape for LSTM (samples, time steps, features)
X_train = X_train.reshape((X_train.shape[0], 1, X_train.shape[1]))
X_val = X_val.reshape((X_val.shape[0], 1, X_val.shape[1]))
X_test = X_test.reshape((X_test.shape[0], 1, X_test.shape[1]))

# Build LSTM Autoencoder
model = keras.Sequential([
    layers.LSTM(16, activation='relu', return_sequences=True, input_shape=(1, 1)),
    layers.LSTM(8, activation='relu', return_sequences=False),
    layers.RepeatVector(1),
    layers.LSTM(8, activation='relu', return_sequences=True),
    layers.LSTM(16, activation='relu', return_sequences=True),
    layers.TimeDistributed(layers.Dense(1))
])

model.compile(optimizer='adam', loss='mse')

# Train model
history = model.fit(X_train, X_train,
                    epochs=50,
                    batch_size=8,
                    validation_data=(X_val, X_val),
                    verbose=1)

# Compute reconstruction error
X_test_pred = model.predict(X_test)
reconstruction_error = np.mean(np.abs(X_test - X_test_pred), axis=1)

# Plot error distribution
plt.hist(reconstruction_error, bins=50)
plt.xlabel("Reconstruction error")
plt.ylabel("Frequency")
plt.title("Anomaly Score Distribution")
plt.show()

# Define anomaly threshold
threshold = np.percentile(reconstruction_error, 95)  # Top 5% as anomalies
anomalies = reconstruction_error > threshold

# Print anomaly detection results
print("Detected anomalies:", sum(anomalies))
```

```python
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers, Model
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt

# Load datasets
train_df = pd.read_csv("train_set.csv")
val_df = pd.read_csv("val_set.csv")
test_df = pd.read_csv("test_set.csv")

# Select numerical feature
X_train = np.array(train_df['num_values']).reshape(-1, 1)
X_val = np.array(val_df['num_values']).reshape(-1, 1)
X_test = np.array(test_df['num_values']).reshape(-1, 1)

# Normalize data
scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_val = scaler.transform(X_val)
X_test = scaler.transform(X_test)

# Reshape for LSTM (samples, time steps, features)
X_train = X_train.reshape((X_train.shape[0], 1, X_train.shape[1]))
X_val = X_val.reshape((X_val.shape[0], 1, X_val.shape[1]))
X_test = X_test.reshape((X_test.shape[0], 1, X_test.shape[1]))

# Build LSTM Autoencoder
inputs = layers.Input(shape=(1, 1))
x = layers.LSTM(16, activation='relu', return_sequences=True)(inputs)
x1 = layers.LSTM(8, activation='relu', return_sequences=False, name="feature_map_layer")(x)
x = layers.RepeatVector(1)(x1)
x = layers.LSTM(8, activation='relu', return_sequences=True)(x)
x = layers.LSTM(16, activation='relu', return_sequences=True)(x)
outputs = layers.TimeDistributed(layers.Dense(1))(x)

model = Model(inputs, outputs)
model.compile(optimizer='adam', loss='mse')

# Train model
history = model.fit(X_train, X_train,
                    epochs=50,
                    batch_size=8,
                    validation_data=(X_val, X_val),
                    verbose=1)

# Compute reconstruction error
X_test_pred = model.predict(X_test)
reconstruction_error = np.mean(np.abs(X_test - X_test_pred), axis=1)

# Plot error distribution
plt.hist(reconstruction_error, bins=50)
plt.xlabel("Reconstruction error")
plt.ylabel("Frequency")
plt.title("Anomaly Score Distribution")
plt.show()

# Define anomaly threshold
threshold = np.percentile(reconstruction_error, 95)  # Top 5% as anomalies
anomalies = reconstruction_error > threshold

# Print anomaly detection results
print("Detected anomalies:", sum(anomalies))

# Feature map visualization
feature_map_model = Model(inputs=model.input, outputs=model.get_layer("feature_map_layer").output)
sample_input = X_test[:1]  # Select one test sample
feature_map = feature_map_model.predict(sample_input)

plt.figure(figsize=(10, 5))
plt.plot(feature_map[0])
plt.title("Feature Map Visualization")
plt.xlabel("Neuron Index")
plt.ylabel("Activation")
plt.show()
```

```python
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers, Model, optimizers
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
import matplotlib.pyplot as plt

# Load datasets
train_df = pd.read_csv("train_set.csv")
val_df = pd.read_csv("val_set.csv")
test_df = pd.read_csv("test_set.csv")

# Select numerical feature
X_train = np.array(train_df['num_values']).reshape(-1, 1)
X_val = np.array(val_df['num_values']).reshape(-1, 1)
X_test = np.array(test_df['num_values']).reshape(-1, 1)

# Normalize data
scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_val = scaler.transform(X_val)
X_test = scaler.transform(X_test)

# Reshape for LSTM (samples, time steps, features)
X_train = X_train.reshape((X_train.shape[0], 1, X_train.shape[1]))
X_val = X_val.reshape((X_val.shape[0], 1, X_val.shape[1]))
X_test = X_test.reshape((X_test.shape[0], 1, X_test.shape[1]))

# Hyperparameter tuning
learning_rate = 0.001
batch_size = 16
num_epochs = 100
optimizer_choice = optimizers.Adam(learning_rate=learning_rate)

# Build LSTM Autoencoder
inputs = layers.Input(shape=(1, 1))
x = layers.LSTM(16, activation='relu', return_sequences=True, trainable=False)(inputs)
x1 = layers.LSTM(8, activation='relu', return_sequences=False, name="feature_map_layer", trainable=False)(x)
x = layers.RepeatVector(1)(x1)
x = layers.LSTM(8, activation='relu', return_sequences=True, trainable=True)(x)
x = layers.LSTM(16, activation='relu', return_sequences=True, trainable=True)(x)
outputs = layers.TimeDistributed(layers.Dense(1))(x)

model = Model(inputs, outputs)
model.compile(optimizer=optimizer_choice, loss='mse')

# Train model with fine-tuning
history = model.fit(X_train, X_train,
                    epochs=num_epochs,
                    batch_size=batch_size,
                    validation_data=(X_val, X_val),
                    verbose=1)

# Compute reconstruction error
X_test_pred = model.predict(X_test)
reconstruction_error = np.mean(np.abs(X_test - X_test_pred), axis=1)

# Define anomaly threshold
threshold = np.percentile(reconstruction_error, 95)  # Top 5% as anomalies
anomalies = reconstruction_error > threshold

# Load true labels
true_labels = np.array(test_df['class'].apply(lambda x: 1 if 'anomaly' in str(x).lower() else 0))

# Evaluate performance
accuracy = accuracy_score(true_labels, anomalies)
precision = precision_score(true_labels, anomalies)
recall = recall_score(true_labels, anomalies)
f1 = f1_score(true_labels, anomalies)
conf_matrix = confusion_matrix(true_labels, anomalies)

# Print evaluation metrics
print(f"Accuracy: {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1-score: {f1:.4f}")
print("Confusion Matrix:")
print(conf_matrix)

# Plot error distribution
plt.hist(reconstruction_error, bins=50)
plt.xlabel("Reconstruction error")
plt.ylabel("Frequency")
plt.title("Anomaly Score Distribution")
plt.show()

# Feature map visualization
feature_map_model = Model(inputs=model.input, outputs=model.get_layer("feature_map_layer").output)
sample_input = X_test[:1]  # Select one test sample
feature_map = feature_map_model.predict(sample_input)

plt.figure(figsize=(10, 5))
plt.plot(feature_map[0])
plt.title("Feature Map Visualization")
plt.xlabel("Neuron Index")
plt.ylabel("Activation")
plt.show()
```
